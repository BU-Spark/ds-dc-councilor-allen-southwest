```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image
API_SECRET_KEY = "############################"
```

# Graphs

```python
tract = 64 
blockGroup = "*"
square = 'B17010_001E,B17010_002E'
title = "Percentage of Families Below Poverty Status w/ Children under 18 Tract {0}, Blocks 1 & 2".format(tract, blockGroup)
df = [["year", "Block Group 1", "Block Group 2"]]
xAxis = df[0][0]
# yAxis = df[0][1]
xLabel = "Year"
yLabel = "Percentage"

for year in range(2010,2022):
    url = 'https://api.census.gov/data/{0}/acs/acs5'.format(year)
    params = {
        'get': square, 
        # 'for': 'block group:{0}'.format(blockGroup),
        'for': 'block group:*',
        'in': 'state:11 county:001 tract:00{0}00'.format(tract),
        'key': API_SECRET_KEY
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        new_data = response.json()
        df.append([year, 100*(int(new_data[1][1])/int(new_data[1][0])), 100*(int(new_data[2][1])/int(new_data[2][0]))])


df = pd.DataFrame(df)
# Set the first row as the column names
df.columns = df.iloc[0]
# Remove the first row
df = df[1:]
# Convert numerical columns to numeric type
df = df.apply(pd.to_numeric, errors='ignore')
# Display the cleaned dataframe
df.reset_index(drop=True, inplace=True)

ax = df.plot.line(x=xAxis, rot=0, ylim=(0, 100))
# modify leged
# ax.get_legend().remove()
ax.legend(title="Block Group")
# levels
ax.set_xlabel(xLabel)
ax.set_ylabel(yLabel)
ax.set_title(title)
print(title)
```
