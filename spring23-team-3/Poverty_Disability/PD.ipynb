```python
list_y = [df_pd_2013,df_pd_2014,df_pd_2015,df_pd_2016,df_pd_2017,df_pd_2018,df_pd_2019,df_pd_2020,df_pd_2021]

# Define a function to modify the label column
def modify_label(label, row_num):
    if row_num == 2:
        label = label + ' (below poverty level)'
    elif row_num in range(3, 9):
        label = label + ' (disability below poverty level)'
    elif row_num == 9:
        label = label + ' (below poverty level)'
    elif row_num in range(10, 16):
        label = label + ' (no disability below poverty level)'
    elif row_num == 17:
        label = label + ' (above poverty level)'
    elif row_num in range(18, 24):
        label = label + ' (disability above poverty level)'
    elif row_num == 24:
        label = label + ' (above poverty level)'
    elif row_num in range(25, 31):
        label = label + ' (no disability above poverty level)'
    return label

# Apply the modify_label function to the Label column
for x in list_y:
    for row_num, label in enumerate(x['Label']):
        x.loc[row_num, 'Label'] = modify_label(label, row_num)


```

```python
# Merge dataframes starting from 2013 to 2021
df_merged = df_pd_2013.merge(df_pd_2014, on='Label').merge(df_pd_2015, on='Label').merge(df_pd_2016, on='Label').merge(df_pd_2017, on='Label').merge(df_pd_2018, on='Label').merge(df_pd_2019, on='Label').merge(df_pd_2020, on='Label').merge(df_pd_2021, on='Label')
```

```python
# Define the list of labels to keep
labels_to_keep = ['Income in the past 12 months below poverty level:', 
                  'With a disability: (below poverty level)',
                  'No disability: (below poverty level)', 
                  'Income in the past 12 months at or above poverty level:', 
                  'With a disability: (above poverty level)', 
                  'No disability: (above poverty level)']

# Create a boolean mask based on the 'Label' column
mask = df_merged['Label'].isin(labels_to_keep)

# Filter the rows of df_merged using the mask
df_merged_new = df_merged[mask]
```
