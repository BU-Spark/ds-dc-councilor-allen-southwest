```python
import pandas as pd
```

```python
df_dp = pd.read_csv('SDC.csv')
```

```python
#Renaming columns
df_dp = df_dp.rename(columns={'Label (Grouping)': 'Labels', 'District of Columbia!!Estimate': 'DC_estimate', 'District of Columbia!!Percent': 'DC_percent'})
```

```python
df_dp.head()
```

Done:

- Changed the column names

To do:

- Replace the labels inside the Labels column where the format '\xa0...' is present
- In the "Labels" column -> put everything into lower case with first letter as upper case if first item is a number.
- Get rid of the values inside DC_percent that are not percentage.
- Get rid of the 'nan' and '(X)' values in all columns.

```python
#Getting rid of the '\xa0\'
#Putting everything into lower character except the first letter

df_dp['Labels'] = df_dp['Labels'].str.lstrip('\xa0')
df_dp['Labels'] = df_dp['Labels'].str.lower().str.capitalize()
```

```python
df_dp["Labels"].unique()
```

```python
import re
```

```python
def replace_non_percentages(value):
    if isinstance(value, str):
        if re.search(r'[^\d\.%]', value):
            return -1
    return value

```

```python
#Get rid of the values inside DC_percent that are not percentage.
df_dp['DC_percent'] = df_dp['DC_percent'].apply(replace_non_percentages)
```

```python
df_dp['DC_percent'].unique()
```

```python
#Replacing the nans in the dataset with -1 
df_dp = df_dp.fillna(-1)
```

```python
df_dp
```

```python
#Replacing the '(X)' values inside the DC_estimate column with -1 
df_dp["DC_estimate"] = df_dp["DC_estimate"].replace('(X)', '-1')
```
