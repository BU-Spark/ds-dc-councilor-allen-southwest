```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x-axis labels
x = ["Tract 64", "Tract 102.01", "Tract 102.02", "Tract 105", "Tract 110.01", "Tract 110.02"]

# Define the y-axis values for whites and location
mobility_whites_within_county = list(map(int,mobility_whites_within_county.values))
mobility_whites_within_state = list(map(int,mobility_whites_within_state.values)) 
total_whites = [mobility_whites_within_county[x] + mobility_whites_within_state[x] for x in range (len (mobility_whites_within_county))]


# Define the y-axis values for black and location
mobility_black_within_county = list(map(float,mobility_black_within_county.values))
mobility_black_within_state = list(map(float,mobility_black_within_state.values)) 
total_blacks = [mobility_black_within_county[x] + mobility_black_within_state[x] for x in range (len (mobility_black_within_county))]

# Define the y-axis values for black and location
mobility_asian_within_county = list(map(float,mobility_asian_within_county.values))
mobility_asian_within_state = list(map(float,mobility_asian_within_state.values)) 
total_asians = [mobility_asian_within_county[x] + mobility_asian_within_state[x] for x in range (len (mobility_asian_within_county))]

# Define the y-axis values for black and location
mobility_hawaiian_within_county = list(map(float,mobility_hawaiian_within_county.values))
mobility_hawaiian_within_state = list(map(float,mobility_hawaiian_within_state.values)) 
total_hawaiians = [mobility_hawaiian_within_county[x] + mobility_hawaiian_within_state[x] for x in range (len (mobility_hawaiian_within_county))]

# Define the y-axis values for black and location
mobility_indian_within_county = list(map(float,mobility_indian_within_county.values))
mobility_indian_within_state = list(map(float,mobility_indian_within_state.values)) 
total_indians = [mobility_indian_within_county[x] + mobility_indian_within_state[x] for x in range (len (mobility_indian_within_county))]

b_asians = list(np.add(total_whites, total_blacks))
b_hawaiians = list(np.add(b_asians, total_asians))
b_indians = list(np.add(b_hawaiians, total_hawaiians))


plt.bar(x, total_whites, 0.7, label="Mobility of Whites", color="#89CFF0")
plt.bar(x, total_blacks, 0.7, bottom= total_whites, label="Mobility of Blacks", color="#6495ED")
plt.bar(x, total_asians, 0.7, bottom=b_asians, label="Mobility of Asians", color="#FFA3B8" )
# plt.bar(x, total_hawaiians, 0.7, bottom=b_hawaiians,label="Mobility of Hawaiians", color="#FB4570")
# plt.bar(x, total_indians, 0.7, bottom=b_indians,label="Mobility of Indians", color="#FF0080")


# Add labels and legends
plt.xticks(x, x, rotation=30, ha='right')
plt.xlabel("Tract")
plt.ylabel("Mobility")
plt.title("Mobility Within County and State for Whites (2016-2021)")
plt.legend()


# Show the chart
plt.show()
```
