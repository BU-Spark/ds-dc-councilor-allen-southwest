```python
# Poverty and Working poor - 2015 - 2020

poverty_status_df = pd.read_csv(PATH + "poverty_2020_15.csv")

poverty_status_full_time = poverty_status_df.iloc[34]
poverty_status_part_time = poverty_status_df.iloc[35]

poverty_status_columns = list(poverty_status_df.columns)


x_axis = []
y_axis1 = []
y_axis2 = []

i = 5
while(i<len(poverty_status_columns)):
    x_axis.append(poverty_status_columns[i].split(',')[0].replace("Census", ""))
    y_axis1.append(float(poverty_status_full_time[i].replace('%', '').replace('-', '0')))
    y_axis2.append(float(poverty_status_part_time[i].replace('%', '').replace('-', '0')))
    i = i+6


import matplotlib.pyplot as plt
import numpy as np
  
# create data
x = np.arange(6)
width = 0.2
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, y_axis1, width, color='#ffd343')
plt.bar(x, y_axis2, width, color='#ff7f0e')
plt.title("Poverty vs Employment Type - 2015-2020")
plt.xticks(x, x_axis, rotation=30, ha='right')
plt.xlabel("Area")
plt.ylabel("Percentage below poverty level")
plt.legend(["Full Time Employees", "Part Time Employees"])
plt.show()
```
