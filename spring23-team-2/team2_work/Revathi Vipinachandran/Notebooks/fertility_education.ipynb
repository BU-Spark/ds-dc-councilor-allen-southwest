```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
PATH = "../data/census/"
fertility_2021_df = pd.read_csv(PATH + "fertility_2021.csv")
fertility_2016_df = pd.read_csv(PATH + "fertility_2016.csv")
fertility_2010_df = pd.read_csv(PATH + "fertility_2010.csv")
```

```python
fertility_2021_df.head(50)
```

```python
i = 5

below_high_school = []
above_high_school = []
associate_degree = []
bachelors_degree = []
graduate_degree = []

while(i<len(fertility_2021_df.columns)):

    below_high_school.append(round(float(fertility_2021_df.iloc[19][i].replace('%','').replace('-', '0'))))
    above_high_school.append(round(float(fertility_2021_df.iloc[20][i].replace('%','').replace('-', '0'))))
    associate_degree.append(round(float(fertility_2021_df.iloc[21][i].replace('%','').replace('-', '0'))))
    bachelors_degree.append(round(float(fertility_2021_df.iloc[22][i].replace('%','').replace('-', '0'))))
    graduate_degree.append(round(float(fertility_2021_df.iloc[23][i].replace('%','').replace('-', '0'))))
    i = i+10


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x-axis labels
x = ["Tract 64", "Tract 102.01", "Tract 102.02", "Tract 105", "Tract 110.01", "Tract 110.02"]

# Define the y-axis values for each gender and location

b_associate_degree = list(np.add(below_high_school, above_high_school))
b_bachelors_degree = list(np.add(b_associate_degree, associate_degree))
b_graduate_degree = list(np.add(b_bachelors_degree, bachelors_degree))


plt.bar(x, below_high_school, 0.7, label="Less than High School", color="#247881")
plt.bar(x, above_high_school, 0.7, bottom=below_high_school, label="High School Graduate", color="#B983FF")
plt.bar(x, associate_degree, 0.7, bottom=b_associate_degree, label="College or Associate Degree", color="#D22779" )
plt.bar(x, bachelors_degree, 0.7, bottom=b_bachelors_degree,label="Bachelors Degree", color="#FF008E")
plt.bar(x, graduate_degree, 0.7, bottom=b_graduate_degree,label="Graduate or Professional Degree", color="#FB4570")

# Add labels and legends
plt.xticks(x, x, rotation=30, ha='right')
plt.xlabel("Tract")
plt.ylabel("Percentage of woman who gave birth")
plt.title("Fertility by Education per Tract (2016-2021)")
plt.legend()

# Loop over each bar and label it with its corresponding value
for i, v in enumerate(below_high_school):
    plt.text(i, v/2, str(v), ha='center', color='white', fontweight='bold', fontsize=8)
    plt.text(i, v+below_high_school[i]/2, str(below_high_school[i]), ha='center', color='white', fontweight='bold', fontsize=8)
    plt.text(i, b_associate_degree[i]+associate_degree[i]/2, str(associate_degree[i]), ha='center', color='black', fontweight='bold', fontsize=8)
    plt.text(i, b_bachelors_degree[i]+bachelors_degree[i]/2, str(bachelors_degree[i]), ha='center', color='white', fontweight='bold', fontsize=8)
    plt.text(i, b_graduate_degree[i]+graduate_degree[i]/2, str(graduate_degree[i]), ha='center', color='white', fontweight='bold', fontsize=8)


# Show the chart
plt.show()
```
