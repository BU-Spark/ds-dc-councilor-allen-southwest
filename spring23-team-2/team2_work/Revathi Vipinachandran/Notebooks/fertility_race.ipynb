```python
i = 5

whites = []
blacks = []
american_indians = []
asians = []
hawaiians = []

while(i<len(fertility_2021_df.columns)):

    whites.append(round(float(fertility_2021_df.iloc[6][i].replace('%','').replace('-', '0'))))
    blacks.append(round(float(fertility_2021_df.iloc[7][i].replace('%','').replace('-', '0'))))
    american_indians.append(round(float(fertility_2021_df.iloc[8][i].replace('%','').replace('-', '0'))))
    asians.append(round(float(fertility_2021_df.iloc[9][i].replace('%','').replace('-', '0'))))
    hawaiians.append(round(float(fertility_2021_df.iloc[10][i].replace('%','').replace('-', '0'))))
    i = i+10

print(whites)
print(blacks)
print(american_indians)
print(asians)
print(hawaiians)

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x-axis labels
x = ["Tract 64", "Tract 102.01", "Tract 102.02", "Tract 105", "Tract 110.01", "Tract 110.02"]

# Define the y-axis values for each gender and location

b_american_indians = list(np.add(whites, blacks))
b_asians = list(np.add(b_american_indians, american_indians))
b_hawaiians = list(np.add(b_asians, asians))


plt.bar(x, whites, 0.7, label="Whites", color="#247881")
plt.bar(x, blacks, 0.7, bottom=whites, label="Blacks", color="#FF6464")
plt.bar(x, american_indians, 0.7, bottom=b_american_indians, label="American Indians", color="#D22779" )
plt.bar(x, asians, 0.7, bottom=b_asians,label="Asians", color="#FF008E")
plt.bar(x, hawaiians, 0.7, bottom=b_hawaiians,label="Hawaiians", color="#FB4570")

# Add labels and legends
plt.xticks(x, x, rotation=30, ha='right')
plt.xlabel("Tract")
plt.ylabel("Percentage of woman who gave birth")
plt.title("Fertility by Race per Tract (2016-2021)")
plt.legend()

# Loop over each bar and label it with its corresponding value
for i, v in enumerate(whites):
    plt.text(i, v/2, str(v), ha='center', color='white', fontweight='bold', fontsize=8)
    plt.text(i, v+blacks[i]/2, str(blacks[i]), ha='center', color='white', fontweight='bold', fontsize=8)
    plt.text(i, b_american_indians[i]+american_indians[i]/2, str(american_indians[i]), ha='center', color='black', fontweight='bold', fontsize=8)
    plt.text(i, b_asians[i]+asians[i]/2, str(asians[i]), ha='center', color='white', fontweight='bold', fontsize=8)
    plt.text(i, b_hawaiians[i]+hawaiians[i]/2, str(hawaiians[i]), ha='center', color='white', fontweight='bold', fontsize=8)


# Show the chart
plt.show()
```
