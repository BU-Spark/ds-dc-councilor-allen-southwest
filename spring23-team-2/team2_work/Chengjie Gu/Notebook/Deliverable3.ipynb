```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import pyplot as plt
PATH = "../Deliverable3/ChengjieData3/"
df = pd.read_csv(PATH+"PopulationChange.csv")
df.dtypes
```

```python
df1 = df[(df["Label (Grouping)"]=="Total Estimate")]
df1
df0 = df1.pivot(index='Year', columns='Tract', values='Black or African American')
df0.plot(title="Black American (population)")
```

```python
df3 = df[(df["Label (Grouping)"]=="Moved from different state	")]
df3
df2 = df1.pivot(index='Year', columns='Tract', values='Black or African American')
df2.plot(title="Black American (percentage moved from different state)")
```
