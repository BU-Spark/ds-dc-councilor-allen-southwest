# Data (Graphs & Infographics)
The following infographics shows several datas related to economic well-being, familty/ community and safety/risky behaviors

```python
import pandas as pd
import matplotlib.pyplot as plt

var = pd.read_excel("Labor Force, Employment, Unemployment ward6.xlsx")

print(var)
var.plot.line(title='Labor Force, Employment, Unemployment in ward6 in DC',x='Year', y=['Labor Force','Employment', 'Unemployment'], grid=True)
var.plot.bar(title='Labor Force, Employment, Unemployment in ward6 in DC',x='Year', y=['Labor Force','Employment', 'Unemployment'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
   
data = {'year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
        'unemployment_rate': [7.3, 7.5, 6.6, 7.3, 6.2, 5.8, 5.1, 5.6, 4.9, 4.8, 7.0],
       }
  
df = pd.DataFrame(data)
print(df)
df.plot(title = 'unemployment_rate in ward6 in DC', x ='year', y='unemployment_rate', kind='line', grid = True)
df.plot(title = 'unemployment_rate in ward6 in DC', x ='year', y='unemployment_rate', kind='bar')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
   
data = {'year': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
        'percentage': [50, 46, 41, 39, 37, 33, 32, 34, 36, 31]
       }
  
df = pd.DataFrame(data)
print(df)
df.plot(title = 'Children in single-parent families in ward6 in DC', x ='year', y='percentage', kind='line')
df.plot(title = 'Children in single-parent families in ward6 in DC', x ='year', y='percentage', kind='bar')
plt.show()
```

Children under age 18 who live with their own single parent either in a family or subfamily.

In this definition, single-parent families may include cohabiting couples and do not include children living with married stepparents. Children who live in group quarters (for example, institutions, dormitories, or group homes) are not included in this calculation.

```python
import pandas as pd
import matplotlib.pyplot as plt
   
data = {'year': [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
        'number': [4000, 3000, 3000, 3000, 2000, 3000, 2000, 2000, 3000]
       }
  
df = pd.DataFrame(data)
print(df)
df.plot(title = 'Teens ages 16 to 19 not attending school and not working in DC', x ='year', y='number', kind='line')
df.plot(title = 'Teens ages 16 to 19 not attending school and not working in DC', x ='year', y='number', kind='bar')
plt.show()
```

Teenagers between age 16 and 19 who are not enrolled in school (full- or part-time) and not employed (full- or part-time).

This measure is sometimes referred to as Idle Teens or Disconnected Youth.
