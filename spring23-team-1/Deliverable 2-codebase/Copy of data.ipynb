```python
import pandas as pd
import numpy as np
from folium import plugins, Map
from folium.plugins import HeatMap
import folium

heatmap = Map()
crimehouse = Map()
gunmap = Map()
gunmapnoheat = Map()


crime = pd.read_csv("Crime_Incidents_in_2021.csv")
crime['LATITUDE'] = crime['LATITUDE'].astype(float)
crime['LONGITUDE'] = crime['LONGITUDE'].astype(float)
crime['WARD'] = crime['WARD'].astype(float)
w6 = crime[crime['WARD'] == 6]
gun = crime[crime['METHOD'] == 'GUN']
c = []
g = []
locations = crime.groupby(['LATITUDE', 'LONGITUDE'], as_index=False)['WARD'].mean()
HeatMap(locations).add_to(heatmap)
heatmap.save("crime.html")
HeatMap(locations).add_to(crimehouse)
HeatMap(locations).add_to(gunmap)

for r in range(len(gun)):
    guns = [gun['LATITUDE'].iloc[r], gun['LONGITUDE'].iloc[r]]
    g.append(guns)

df = pd.read_csv("Affordable_Housing.csv")
base_map = Map()

df['LATITUDE'] = df['LATITUDE'].astype(float)
df['LONGITUDE'] = df['LONGITUDE'].astype(float)

ward6 = df[df['MAR_WARD'] == 'Ward 6']
notward6 = df[df['MAR_WARD'] != 'Ward 6']

ward6location = []
not6location = []

for r in range(len(ward6)):
    location = [ward6['LATITUDE'].iloc[r], ward6['LONGITUDE'].iloc[r]]
    ward6location.append(location)

for r in range(len(notward6)):
    location = [notward6['LATITUDE'].iloc[r], notward6['LONGITUDE'].iloc[r]]
    not6location.append(location)


for i in ward6location:
    folium.CircleMarker(i, radius=5, popup='Ward 6',color = 'blue',fill_color='blue').add_to(base_map)
    folium.CircleMarker(i, radius=5, popup='Ward 6',color = 'blue',fill_color='blue').add_to(crimehouse)
    folium.CircleMarker(i, radius=5, popup='Ward 6',color = 'blue',fill_color='blue').add_to(gunmap)
    folium.CircleMarker(i, radius=5, popup='Ward 6',color = 'blue',fill_color='blue').add_to(gunmapnoheat)

for i in not6location:
    folium.CircleMarker(i, radius=5,popup='Not Ward 6', color = 'green', fill_color='green').add_to(base_map)
    folium.CircleMarker(i, radius=5,popup='Not Ward 6', color = 'green', fill_color='green').add_to(crimehouse)
    folium.CircleMarker(i, radius=5,popup='Not Ward 6', color = 'green', fill_color='green').add_to(gunmap)
    folium.CircleMarker(i, radius=5,popup='Not Ward 6', color = 'green', fill_color='green').add_to(gunmapnoheat)


for i in g:
    folium.CircleMarker(i, radius=5,popup='Gun Violence', color = 'red', fill_color='red').add_to(gunmap)
    folium.CircleMarker(i, radius=5,popup='Gun Violence', color = 'red', fill_color='red').add_to(gunmapnoheat)


base_map.save("affordablehousing.html")
crimehouse.save("HousingWithCrime.html")
gunmap.save("GunViolenceHeat.html")
gunmapnoheat.save("GunViolence.html")

```
