```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("The Below graphs shows the distribution of poverty data over different tracts. ")
def renderPovertyData(excelName, sheetName, tractName):
    df = pd.read_excel(excelName, sheet_name=sheetName)
    df = df.dropna()
    plt.figure().set_figwidth(20)
    plt.xticks(rotation = 90) 
    plt.plot(df["Label (Grouping)"], df["Total"], label = "Total")
    plt.plot(df["Label (Grouping)"], df["Below poverty level"], label = "Below poverty level")
    plt.title('Poverty Comaprision in ' + tractName, fontsize=10)
    plt.xlabel('Grouping', fontsize=10)
    plt.ylabel('Count', fontsize=10)
    plt.legend()
    plt.ion()
    plt.show()

excelName = "PovertyStatus.xlsx"
sheetNames = [
    "DC",
    "Tract64", 
    "Tract102.01",
    "Tract102.02",
    "Tract105",
    "Tract110.01",
    "Tract110.02"]

for sheetName in sheetNames:
    renderPovertyData(excelName, sheetName, sheetName)

print("The above graphs shows that tract 64 has a lot more individuals who are between the age of 18 to 64 years who are below the poverty level. \nTackling this problem or analysing this problem can help us understand more about the issues with violance and crime in DC ")
```

```python
print("Below are the grpahs which depict the poverty distribution according to Race.")

def renderRacePovertyData(excelName, sheetName, tractName):
    df = pd.read_excel(excelName, sheet_name=sheetName)
    df = df.dropna()
    fig, ax = plt.subplots()

    ind = np.arange(len(df))
    width = 0.4
    ax.barh(ind, df["Total"], label='Total')
    ax.barh(ind, df["Below poverty level"], label='Below poverty level')
    ax.set(yticks=ind, yticklabels=df["Label (Grouping)"])

    # df = pd.DataFrame({'Total': df["Total"], 'Below Poverty Level': df["Below poverty level"]}, index=df["Label (Grouping)"])
    # ax = df.plot.barh()

    plt.title('Race Poverty Comaprision in ' + tractName, fontsize=10)
    plt.ylabel('Race', fontsize=10)
    plt.xlabel('Count', fontsize=10)
    ax.legend()
    plt.show()

excelName = "RacePovertyStatus.xlsx"
sheetNames = [
    "DC",
    "Tract64", 
    "Tract102.01",
    "Tract102.02",
    "Tract105",
    "Tract110.01",
    "Tract110.02"]

for sheetName in sheetNames:
    renderRacePovertyData(excelName, sheetName, sheetName)
```
