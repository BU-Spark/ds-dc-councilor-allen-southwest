```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

crime = pd.read_csv("dc-crimes-search-results.csv")
```

```python
graphDf = crime.groupby(["offensegroup", "offense-text"]).count()
```

```python
df = crime[:]
pd.crosstab(df["offensegroup"], df["offense-text"]).plot.barh()

print("The below graph shows the number of different property and violent crime which have occured in the whole of Washington D.C. \nThis clearly shows that the number of property crimes are more in Washington D.C. when compared to violent crimes.")
```

```python
df = crime[crime['WARD'] == 6.0]
pd.crosstab(df["offensegroup"], df["offense-text"]).plot.barh()

print("The below graph shows the similar data which is available in DC Crime Cards but it is concentrated on Ward 6. \nBased the two graphs, we can see that both the graphs have similar proportions of violent and property crimes")
```

```python
df = crime.groupby(["offensegroup"]).count()

offense = df["offensekey"].keys()

data = df["offense-text"].array

# Creating explode data
explode = (0.1, 0.0)
 
# Creating color parameters
colors = ( "lightblue",
          "yellow", "light blue", )
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} incidents)".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = offense,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="black"))
 
# Adding legend
ax.legend(wedges, offense,
          title ="Offense Categories",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Number of crimes by Offense Categories")
 
# show plot
plt.show()

print("The above pie chart shows that about 85.9% of the total recorded crimes in the past two years which amounts upto 48979 incidents are property crimes.")
```

```python
df = crime.groupby(["WARD"]).count()

ward = df["DISTRICT"].keys()
data = df["DISTRICT"].array
 
# Creating explode data
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.2, 0.05, 0.05)
 
# Creating color parameters
colors = ( "blue",
          "yellow", "lightblue", "white", "red", "lightgreen", "gray", "violet")
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} incidents)".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = ward,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 270,
                                  wedgeprops = wp,
                                  textprops = dict(color ="black"))
 
# Adding legend
ax.legend(wedges, ward,
          title ="Wards",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Number of crimes by Wards in DC")
 
# show plot
plt.show()

print("From the above pie chart we can clearly see that the crimes in Ward 6 accounts upto around 13.4% of the total crimes in Washington D.C.")
```
