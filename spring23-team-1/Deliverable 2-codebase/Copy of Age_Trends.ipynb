```python
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt


info = "Considering the total population from past 5 years the below graph is ploted, It shows that Tract 102.02 is the most populated Tract in the past 5 years"
print(info)
print("\n")
# Data for the plot
years = [
    "Tract64", 
    "Tract102.01",
    "Tract102.02",
    "Tract105",
    "Tract110.01",
    "Tract110.02"]
populations = [2522,1960,4027,2385,1519,2492]

# Create a bar plot
plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(8,5))
ax.bar(years, populations, color=['lightblue'])

# Set the plot title and axis labels
ax.set_title("Total Population Tract wise - ACS 5 Years", transform=fig.transFigure, ha='center', fontsize=14, weight='bold', alpha=.8)
ax.set_xlabel("Year")
ax.set_ylabel("Total Population")

# ax.text(x=0.12, y=.93, s="Average Airlines Delay per Month in 2008", transform=fig.transFigure, ha='left', fontsize=14, weight='bold', alpha=.8)
# ax.text(x=0.12, y=.90, s="Difference in minutes between scheduled and actual arrival time averaged over each month", transform=fig.transFigure, ha='left', fontsize=12, alpha=.8)

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def renderIncomeData(excelName, sheetName, tractName):
    df = pd.read_excel(excelName, sheet_name=sheetName)
    df = df.dropna()
    
    fig, ax = plt.subplots(figsize=(20,20))

    ind = np.arange(len(df))
    width = 0.4
    ax.barh(ind, df["Total"], label = "Total")
    ax.barh(ind, df["Male"], label = "Male")
    ax.barh(ind, df["Female"], label = "Female")
#   ax.barh(ind, df["Nonfamily households"],label = "Nonfamily households")
    ax.set(yticks=ind, yticklabels=df["Label"])

    # df = pd.DataFrame({'Total': df["Total"], 'Below Poverty Level': df["Below poverty level"]}, index=df["Label (Grouping)"])
    # ax = df.plot.barh()

    plt.title('Population Trends for different ages ' + tractName, fontsize=14, weight='bold', alpha=.8)
    plt.xlabel('Grouping', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    ax.legend()
    plt.show()

excelName = "ageTrends_Tractwise.xlsx"
sheetNames = [
    "DC",
    "Tract64", 
    "Tract102.01",
    "Tract102.02",
    "Tract105",
    "Tract110.01",
    "Tract110.02"]

for sheetName in sheetNames:
    renderIncomeData(excelName, sheetName, sheetName)
```
