```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def renderIncomeData(excelName, sheetName, tractName):
    df = pd.read_excel(excelName, sheet_name=sheetName)
    df = df.dropna()  
    fig, ax = plt.subplots()

    ind = np.arange(len(df))
    width = 0.4
    ax.barh(ind, df["Households"], label = "Households")
    ax.barh(ind, df["Families"], label = "Families")
    ax.barh(ind, df["Married-couple families"], label = "Married-couple families")
    ax.barh(ind, df["Nonfamily households"],label = "Nonfamily households")
    ax.set(yticks=ind, yticklabels=df["Label"], ylim=[2*width - 1, len(df)])

    # df = pd.DataFrame({'Total': df["Total"], 'Below Poverty Level': df["Below poverty level"]}, index=df["Label (Grouping)"])
    # ax = df.plot.barh()

    plt.title('Poverty Comaprision in ' + tractName, fontsize=10)
    plt.xlabel('Grouping', fontsize=10)
    plt.ylabel('Count', fontsize=10)
    ax.legend()
    plt.show()

excelName = "IncomePastYear copy.xlsx"
sheetNames = [
    "DC",
    "Tract64", 
    "Tract102.01",
    "Tract102.02",
    "Tract105",
    "Tract110.01",
    "Tract110.02"
]

for sheetName in sheetNames:
    renderIncomeData(excelName, sheetName, sheetName)
```
